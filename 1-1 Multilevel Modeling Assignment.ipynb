```r
# Importing the necessary packages
library(ggplot2)
library(nlme)
library(lme4)
```

```r
# Importing the necessary dataset
df_popular <- read.csv('0-2 df_pupil.csv')
```

```r
# Removing the index column from the data fram
df_popular <- subset(df_popular, select = -X)
```

```r
# Examining the structure of the data frame
str(df_popular) 
```

```r
# Modifying the nature of the 'class' variable
df_popular$extravXsex <- df_popular$extrav * df_popular$sex
df_popular$class <- as.factor(df_popular$class)
df_popular$sex <- as.factor(df_popular$sex)
```

```r
# Formulating justifications for multilevel modeling

# 1. Checking whether intercepts and slopes vary by group
ggplot(df_popular, aes(x = popular, y = extrav, color = class)) + geom_point() + geom_smooth(method = 'lm', se = FALSE)
```

```r
# Calculating the null model
nullModel <- lme(popular ~ 1, data = df_popular, random = ~ 1|class, method = 'ML')
summary(nullModel)
```

```r
VarCorr(nullModel)
```

```r
# Calculating vector of variance estimates
varests <- as.numeric(VarCorr(nullModel)[1:2])

# Calculating ICC
ICC <- varests[1]/sum(varests)
ICC
```

```r
# Calculating a model with only level 1 predictors
ctrl <- lmeControl(opt='optim')
model1 <- lme(popular ~ extrav + sex + extrav*sex, data = df_popular, random = ~ extrav + sex + extrav*sex| class, method = 'ML', control = ctrl)
summary(model1)
```

```r
# Examining the variance components in the model
VarCorr(model1)
```

```r
# Creating two more models, which are:
# 1. Model with random intercepts (model2)
# 2. Model with random intercepts and slopes (model3)

model2 <- lme(popular ~ extrav + sex + texp + extrav*sex, data = df_popular, random = ~ extrav + sex + extrav*sex| class, method = 'ML', control = ctrl)
model3 <- lme(popular ~ extrav + sex + texp + extrav*sex + texp*extrav + texp*sex, data = df_popular, random = ~ extrav + sex + extrav*sex| class, method = 'ML', control = ctrl)
```

```r
summary(model2)
```

```r
summary(model3)
```

```r
# Menentukan model1, model2, dan model3 tanpa interaksi antara extrav dan sex
model4 <- lme(popular ~ extrav + sex, data = df_popular, random = ~ extrav + sex | class, method = 'ML', control = ctrl)
model5 <- lme(popular ~ extrav + sex + texp, data = df_popular, random = ~ extrav + sex | class, method = 'ML', control = ctrl)
model6 <- lme(popular ~ extrav + sex + texp + texp*extrav + texp*sex, data = df_popular, random = ~ extrav + sex | class, method = 'ML', control = ctrl)
```

```r
summary(model4)
```

```r
summary(model5)
```

```r
summary(model6)
```

```r
# Mencoba model tanpa texp*sex sebagai prediktor
model7 <- lme(popular ~ extrav + sex + texp + extrav*sex + texp*extrav, data = df_popular, random = ~ extrav + sex + extrav*sex| class, method = 'ML', control = ctrl)
model8 <- lme(popular ~ extrav + sex + texp + texp*extrav, data = df_popular, random = ~ extrav + sex | class, method = 'ML', control = ctrl)
```
